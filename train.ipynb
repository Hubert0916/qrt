```python
from models.quantile_regression_tree import QuantileRegressionTree
from quantile_forest import RandomForestQuantileRegressor
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import altair as alt
```

```python
df = pd.read_csv(f"data/esg_tfidf_with_return2.csv", encoding='utf-8-sig').head(500)
x_cols = [c for c in df.columns if c.endswith('_TFIDF')]
x = df[x_cols]
y = df["報酬率"]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
df_test = df.iloc[y_test.index]
```

```python
qrf = RandomForestQuantileRegressor(max_depth=5, min_samples_leaf=5)
qrf.fit(X_train, y_train)
```

```python
qrt = QuantileRegressionTree(max_depth=5, min_samples_leaf=5, feature_names=x_cols)
qrt.fit(X_train, y_train, quantile=0.7)
```

```python
y_qrf_70 = qrf.predict(X_test, quantiles=0.7)

y_qrt_70 = qrt.predict(X_test)
```

```python
df_test['qrf_70'] = y_qrf_70
df_test['qrt_70'] = y_qrt_70
```
